# *Matemática Computacional* - ***Atividade 0***


*O vigente notebook tem como objetivo aplicar as matrizes tranformação de
rotação e shear (cisalhamento) em vetores arbitrários.*

Com isso, utilizaremos as seguintes bibliotecas:
* `numpy` para executar as operações matemáticas e criação de arrays.
* `matplotlib` para desenhar os gráficos que serão utilizados.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### **Função Para Desenhar o Gráfico**

```python
colors = ['#fc3468', '#e403dc']

def plot_graph(list_vet, name_graph):

  plt.figure(figsize = (6,6))

  plt.title(name_graph)

  #confingurando os intervalos do gráfico
  plt.axis([-5,5,-5,5])

  plt.xticks(range(-5, 6))
  plt.yticks(range(-5, 6))

  #definido as linhas horizontais e verticais do gráfico
  plt.axhline(0)
  plt.axvline(0)

  #iterando para desenhar cada reta
  for i, vet in enumerate(list_vet):
    plt.arrow(0,0,vet[0], vet[1], head_width=0.2, head_length=0.2, fc=colors[i % len(colors)], ec=colors[i % len(colors)], length_includes_head = True, zorder = 3)
    position_x = vet[0] + 0.2
    position_y = vet[1] + 0.2 if vet[0] >= 0 else vet[0] - 0.2
    plt.text(position_x, position_y, f"{int(vet[0]), int(vet[1])}", color=colors[i])

  plt.grid()
  plt.show()
```

### **Função Produto** - *Transformação Linear*

```python
#a matriz multiplica cada vetor para obtermos a rotação desejada
def dot(vet1, vet2, matriz):
  return np.dot(matriz, vet1), np.dot(matriz, vet2)
```

### **Rotação em 90°** - *Vetores Unitários*

Para fazer a rotação em 90° devemos escolher dois vetores unitários, deste modo temos os seguintes vetores base:

$$
\hat{i} = \begin{pmatrix}
1 \\
0
\end{pmatrix}
$$

$$
\hat{j} = \begin{pmatrix}
0 \\
1
\end{pmatrix}
$$

Inserindo os valores dos vetores no gráfico com a função `plot_graph` , teremos:

```python
plot_graph([[1,0], [0,1]], 'VETORES UNITÁRIOS')
```

Sendo assim, quando os vetores $\hat{i}$ e $\hat{j}$ sofrerem a rotação, teremos os novos vetores que irão representar a base do espaço transformado:

$$
\hat{i} = \begin{pmatrix}
0 \\
-1
\end{pmatrix}
$$

$$
\hat{j} = \begin{pmatrix}
1 \\
0
\end{pmatrix}
$$

```python
plot_graph([[0,-1],[1,0]], 'BASE DO ESPAÇO')
```

Logo, a partir das bases que obtemos do espaço transformado podemos montar a matriz rotação que será utilizada para transformar um vetor arbitrário:

$$
\ M_B= \begin{pmatrix}
0 & 1\\
-1 & 0
\end{pmatrix}
$$

----

### **Rotação em 90°** - *Transformação Linear*

Com isso, para rotacionarmos dois vetores arbitrários usaremos a base obtida e deste modo faremos uma transformação linear no espaço.
Os vetores escolhidos são:

$$
\ V_A= \begin{pmatrix}
1 \\
1
\end{pmatrix}
$$

$$
\ V_B= \begin{pmatrix}
1 \\
-1
\end{pmatrix}
$$

Desenhado no gráfico:

```python
plot_graph([[1,1],[1,-1]], 'VETORES ARBITRÁRIOS')
```

Aplicando a transformação linear, temos então o novo gráfico:

```python
#os vetores 1 e 2 recebem os novos valores das retas
vetor1, vetor2 = dot([1,1],[1,-1], [[0,1],[-1,0]])

plot_graph([vetor1, vetor2], 'ROTAÇÃO 90°')

```

### **Shear (Cisalhamento)** - *Vetores Unitários*

O conceito de cisalhamento define-o como: Uma transformação linear que inclina uma determinada figura, essa transformação pode tanto no eixo x quanto no eixo y.

Com isso, temos os seguintes vetores:

$$
\hat{i} = \begin{pmatrix}
1 \\
0
\end{pmatrix}
$$

$$
\hat{j} = \begin{pmatrix}
0 \\
1
\end{pmatrix}
$$

```python
plot_graph([[1,0],[0,1]], 'VETORES UNITÁRIOS')
```

Aplicando um fator de cisalhamento 2 na horizontal, ou seja, alterando o x do vetor $\hat{j}$, teremos os seguintes vetores:

$$
\hat{i} = \begin{pmatrix}
1 \\
0
\end{pmatrix}
$$

$$
\hat{j} = \begin{pmatrix}
2 \\
1
\end{pmatrix}
$$

Desenhando o gráfico:

```python
plot_graph([[1,0],[2,1]], 'FATOR DE CISALHAMENTO')
```

A partir do gráfico, temos que a matriz base para o cisalhamento será:

$$
M_B = \begin{pmatrix}
1 & 2 \\
0 & 1
\end{pmatrix}
$$

----

### **Shear (Cisalhamento)** - *Transformação Linear*

Sendo assim, podemos utilizar a matriz $M_B$ para aplicar o cisalhamento em vetores arbitrários.  

Os vetores escolhidos são:

$$
\ V_A= \begin{pmatrix}
1 \\
1
\end{pmatrix}
$$

$$
\ V_B= \begin{pmatrix}
1 \\
-1
\end{pmatrix}
$$

```python
plot_graph([[1,1], [1,-1]], 'VETORES ARBITRÁRIOS')
```

Aplicando agora o cisalhamento nos dois vetores:

```python
#os vetores 1 e 2 recebem os novos valores das retas
vetor1, vetor2 = dot([1,1], [1,-1], [[1,2],[0,1]])

plot_graph([vetor1, vetor2], 'CISALHAMENTO')
```
